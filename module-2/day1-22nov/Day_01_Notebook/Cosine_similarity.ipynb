```python


# Import necessary libraries
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Load a sentence embedding model
model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# Define sentences
# sentences = [
#     "The cat is sitting on the mat.",
#     "A dog is resting on the carpet.",
#     "The quick brown fox jumps over the lazy dog.",
#     "The weather is sunny and warm today.",
#     "It's a bright and beautiful day outside."
# ]
```

```python
sentences =["kill two birds with one stone.",

            "Achieve two goals with a single action"]
```

```python



# Function to calculate sentence embeddings
def get_sentence_embedding(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    # Average the token embeddings to get a sentence-level embedding
    sentence_embedding = outputs.last_hidden_state.mean(dim=1)
    return sentence_embedding

# Calculate embeddings for all sentences
embeddings = [get_sentence_embedding(sentence) for sentence in sentences]


```

```python
import numpy as np

def cosine_similarity(embedding1, embedding2):
    # Dot product of the two embeddings
    dot_product = np.dot(embedding1, embedding2)

    # Norm (magnitude) of each embedding
    norm_embedding1 = np.linalg.norm(embedding1)
    norm_embedding2 = np.linalg.norm(embedding2)

    # Cosine similarity formula
    similarity = dot_product / (norm_embedding1 * norm_embedding2)

    return similarity
```

```python
embedding1 = embeddings[0].flatten()
embedding2 = embeddings[1].flatten()
```
