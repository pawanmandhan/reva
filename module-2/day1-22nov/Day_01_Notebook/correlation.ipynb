```python
def correlation(col1,col2):
  mean1=np.mean(col1)
  mean2=np.mean(col2)
  new_col1=col1-mean1
  new_col2=col2-mean2
  numerator =np.dot(new_col1,new_col2)
  denominator= np.linalg.norm(new_col1)*np.linalg.norm(new_col2)
  return numerator/denominator
```

```python
correlation(data["mpg"],data["weight"])
```

```python
correlation(data["cylinders"],data["weight"])
```
