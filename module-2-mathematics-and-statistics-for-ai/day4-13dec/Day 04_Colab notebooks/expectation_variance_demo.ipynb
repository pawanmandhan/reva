# Understanding Expectation and Variance Through Simulation



This notebook demonstrates the concepts of **expected value**, **variance**, and the **law of large numbers** using simulation.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## ðŸŽ² Expected Value of a Fair Die
Theoretical expectation:
\(\mathbb{E}[X] = 3.5\)


```python
def simulate_die(n):
    rolls = np.random.randint(1,7,size=n)
    return rolls.mean()

for n in [10,100,1000,100000]:
    print(f"Average of {n} rolls: {simulate_die(n):.4f}")
```

##  Law of Large Numbers

As the number of independent repetitions of a random experiment increases, the sample average gets closer and closer to the theoretical expected value.

```python
N=10000
rolls=np.random.randint(1,7,size=N)
running_avg=np.cumsum(rolls)/np.arange(1,N+1)
plt.figure(figsize=(10,5))
plt.plot(running_avg)
plt.axhline(3.5,color='red',linestyle='--')
plt.show()
```

##  Variance of a Fair Die

```python
rolls=np.random.randint(1,7,size=100000)
print('Empirical variance:',rolls.var())
print('Theoretical variance: 2.9167')
```

## Histogram of Die Rolls

```python
plt.hist(rolls,bins=np.arange(1,8)-0.5,edgecolor='black')
plt.show()
```

## Binomial Example: Number of Heads in 5 Tosses

```python
def simulate_heads(n):
    flips=np.random.randint(0,2,size=(n,5))
    return flips.sum(axis=1)

heads=simulate_heads(100000)
print('Empirical mean:',heads.mean())
print('Empirical variance:',heads.var())
```

## Histogram of Heads Distribution

```python
plt.hist(heads,bins=np.arange(-0.5,6.5,1),edgecolor='black')
plt.show()
```
