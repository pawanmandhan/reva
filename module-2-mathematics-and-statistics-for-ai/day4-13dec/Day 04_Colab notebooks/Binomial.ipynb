###  Binomial Probability Problem – Drug-Related Thefts

In a certain city district, the need for money to buy drugs is stated as the reason for **75%** of all thefts.

Let the number of drug-related thefts among upcoming cases be modeled using a **Binomial distribution** with:

- Probability of drug-related theft: **p = 0.75**
- Number of theft cases considered: **n = 5**

Answer the following:

---

#### **(a)**  
What is the probability that **exactly 2** of the next 5 theft cases resulted from the need for money to buy drugs?

---

#### **(b)**  
What is the probability that **at most 3** of the next 5 theft cases resulted from the need for money to buy drugs?

---


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import comb

# Parameters
p = 0.75
q = 1 - p
n = 5

# Compute binomial PMF values
k_values = np.arange(0, n+1)
pmf_values = np.array([comb(n, k) * (p**k) * (q**(n-k)) for k in k_values])

# Create a probability table
prob_table = pd.DataFrame({
    "k (Number of drug-related thefts)": k_values,
    "P(X = k)": pmf_values
})

prob_table
```

```python
plt.figure(figsize=(8,5))
plt.bar(k_values, pmf_values)
plt.xlabel("Number of Drug-Related Thefts (k)")
plt.ylabel("P(X = k)")
plt.title("Binomial PMF: n = 5, p = 0.75")
plt.grid(True, axis='y')
plt.show()
```

```python
# (a) P(X = 2)
p_a = comb(n, 2) * (p**2) * (q**3)

# (b) P(X <= 3)

p_b = 0
for k in range(0, 4):      # k = 0, 1, 2, 3
    p_b += comb(n, k) * (p**k) * (q**(n - k))


print("RESULTS")
print("---------------------------")
print(f"(a) P(X = 2) = {p_a:.6f}")
print(f"(b) P(X <= 3) = {p_b:.6f}")
```

###  Binomial Probability Problem – Pipework Failures

According to *Chemical Engineering Progress* (November 1990), approximately **30%** of all pipework failures in chemical plants are caused by **operator error**.

Assume that the number of operator-error failures among upcoming cases follows a **Binomial distribution** with:

- Probability of operator error: **p = 0.30**
- Number of failures considered: **n = 20**

Answer the following:

---

#### **(a)**  
What is the probability that **at least 10** out of the next 20 pipework failures are due to operator error?

---

#### **(b)**  
What is the probability that **no more than 4** out of 20 such failures are due to operator error?

---


```python
from math import comb

p = 0.30
q = 1 - p
n = 20

# (a) P(X >= 10)
p_a = 0
for k in range(10, n+1):  # 10 to 20
    p_a += comb(n, k) * (p**k) * (q**(n-k))

# (b) P(X <= 4)
p_b = 0
for k in range(0, 5):  # 0 to 4
    p_b += comb(n, k) * (p**k) * (q**(n-k))


print("RESULTS")
print("------------------------------")
print(f"(a) P(X ≥ 10) = {p_a:.6f}")
print(f"(b) P(X ≤ 4)  = {p_b:.6f}")
```
