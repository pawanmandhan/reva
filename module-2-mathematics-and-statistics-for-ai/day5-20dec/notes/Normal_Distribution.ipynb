```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
from matplotlib.patches import Rectangle

plt.style.use('seaborn-v0_8-darkgrid')
print("Libraries imported successfully!")
```

## Section 2: What is the Normal Distribution?

### Definition:
The **Normal Distribution** is a continuous probability distribution defined by two parameters:
- **Mean (μ)**: Center of the distribution
- **Standard Deviation (σ)**: Spread/width of the distribution

### Probability Density Function (PDF):
$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

Where:
- $x$ = value
- $\mu$ = population mean
- $\sigma$ = population standard deviation
- $e$ and $\pi$ are mathematical constants

### Key Characteristics:
1. **Bell-shaped curve** - Symmetric around the mean
2. **Unimodal** - Single peak at the mean
3. **Mean = Median = Mode** - All at the same point
4. **Asymptotic** - Tails never touch the x-axis
5. **Empirical Rule** - Specific percentages within standard deviations

## Section 3: The Empirical Rule (68-95-99.7 Rule)

```python
# Visualize the Empirical Rule
fig, ax = plt.subplots(figsize=(14, 8))

mu = 100
sigma = 15
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

# Plot the normal distribution
ax.plot(x, y, 'b-', linewidth=2.5, label='Normal Distribution')
ax.fill_between(x, y, alpha=0.1, color='blue')

# Shade regions for empirical rule
colors = ['lightgreen', 'lightyellow', 'lightcoral']
labels = ['68%', '95%', '99.7%']
sigmas = [1, 2, 3]

for sigma_mult, color, label in zip(sigmas, colors, labels):
    x_fill = x[(x >= mu - sigma_mult*sigma) & (x <= mu + sigma_mult*sigma)]
    y_fill = stats.norm.pdf(x_fill, mu, sigma)
    ax.fill_between(x_fill, y_fill, alpha=0.3, color=color, label=f'±{sigma_mult}σ: {label} of data')

# Add vertical lines at mean and standard deviations
ax.axvline(mu, color='black', linestyle='-', linewidth=2, label=f'Mean (μ = {mu})')

for i in range(1, 4):
    ax.axvline(mu + i*sigma, color='gray', linestyle='--', linewidth=1, alpha=0.7)
    ax.axvline(mu - i*sigma, color='gray', linestyle='--', linewidth=1, alpha=0.7)

# Add text annotations
ax.text(mu, 0.015, f'μ', fontsize=12, ha='center', fontweight='bold')
ax.text(mu + sigma, 0.018, f'+σ', fontsize=10, ha='center')
ax.text(mu - sigma, 0.018, f'-σ', fontsize=10, ha='center')
ax.text(mu + 2*sigma, 0.015, f'+2σ', fontsize=10, ha='center')
ax.text(mu - 2*sigma, 0.015, f'-2σ', fontsize=10, ha='center')
ax.text(mu + 3*sigma, 0.012, f'+3σ', fontsize=10, ha='center')
ax.text(mu - 3*sigma, 0.012, f'-3σ', fontsize=10, ha='center')

ax.set_xlabel('Value', fontsize=12, fontweight='bold')
ax.set_ylabel('Probability Density', fontsize=12, fontweight='bold')
ax.set_title('The Empirical Rule: 68-95-99.7', fontsize=14, fontweight='bold')
ax.legend(loc='upper right', fontsize=10)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('empirical_rule.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n" + "="*70)
print("THE EMPIRICAL RULE (68-95-99.7 Rule)")
print("="*70)
print("""
For a normal distribution:

1. ±1σ (Within 1 standard deviation of mean):
   - Contains approximately 68% of the data
   - Range: μ - σ to μ + σ

2. ±2σ (Within 2 standard deviations of mean):
   - Contains approximately 95% of the data
   - Range: μ - 2σ to μ + 2σ

3. ±3σ (Within 3 standard deviations of mean):
   - Contains approximately 99.7% of the data
   - Range: μ - 3σ to μ + 3σ

This means:
- 32% of data falls outside ±1σ
- 5% of data falls outside ±2σ
- 0.3% of data falls outside ±3σ (very rare)
""")
print("="*70)
```

## Section 4: The Standard Normal Distribution (Z-Distribution)

```python
print("\n" + "="*70)
print("STANDARD NORMAL DISTRIBUTION (Z-Distribution)")
print("="*70)
print("""
The Standard Normal Distribution is a special case of the normal distribution:

Properties:
  • Mean (μ) = 0
  • Standard Deviation (σ) = 1
  • Notation: N(0, 1)

Why is it important?
  • Simplifies calculations
  • Used as reference for all normal distributions
  • Foundation for Z-tests and Z-scores
  • Any normal distribution can be converted to standard normal

Conversion Formula (Z-Score):
  Z = (X - μ) / σ
  
Where:
  • Z = Z-score (standardized value)
  • X = original value
  • μ = population mean
  • σ = population standard deviation

Interpretation of Z-score:
  • Z = 0: Value equals the mean
  • Z = 1: Value is 1 standard deviation above mean
  • Z = -2: Value is 2 standard deviations below mean
  • Z > 3 or Z < -3: Value is very rare (outlier)
""")
print("="*70)
```

## Section 5: Visualizing Different Normal Distributions

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

x = np.linspace(-10, 20, 1000)

# Plot 1: Different means, same standard deviation
ax = axes[0, 0]
means = [0, 5, 10]
colors = ['red', 'green', 'blue']

for mean, color in zip(means, colors):
    y = stats.norm.pdf(x, mean, 2)
    ax.plot(x, y, color=color, linewidth=2.5, label=f'μ={mean}, σ=2')

ax.set_title('Effect of Different Means (σ fixed)', fontsize=12, fontweight='bold')
ax.set_xlabel('Value')
ax.set_ylabel('Probability Density')
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

# Plot 2: Same mean, different standard deviations
ax = axes[0, 1]
stds = [1, 2, 3]
colors = ['red', 'green', 'blue']

for std, color in zip(stds, colors):
    y = stats.norm.pdf(x, 5, std)
    ax.plot(x, y, color=color, linewidth=2.5, label=f'μ=5, σ={std}')

ax.set_title('Effect of Different Standard Deviations (μ fixed)', fontsize=12, fontweight='bold')
ax.set_xlabel('Value')
ax.set_ylabel('Probability Density')
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

# Plot 3: Standard normal distribution
ax = axes[1, 0]
x_std = np.linspace(-4, 4, 1000)
y_std = stats.norm.pdf(x_std, 0, 1)

ax.plot(x_std, y_std, 'b-', linewidth=2.5, label='Standard Normal N(0,1)')
ax.fill_between(x_std, y_std, alpha=0.2, color='blue')
ax.axvline(0, color='red', linestyle='--', linewidth=2, label='Mean = 0')

# Shade regions
x_fill = x_std[np.abs(x_std) <= 1]
y_fill = stats.norm.pdf(x_fill, 0, 1)
ax.fill_between(x_fill, y_fill, alpha=0.3, color='green', label='±1σ (68%)')

ax.set_title('Standard Normal Distribution', fontsize=12, fontweight='bold')
ax.set_xlabel('Z-Score')
ax.set_ylabel('Probability Density')
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

# Plot 4: Cumulative distribution function
ax = axes[1, 1]
y_cdf = stats.norm.cdf(x_std, 0, 1)

ax.plot(x_std, y_cdf, 'b-', linewidth=2.5, label='CDF')
ax.axhline(0.5, color='red', linestyle='--', linewidth=1.5, alpha=0.7, label='P(Z ≤ 0) = 0.5')
ax.axvline(0, color='red', linestyle='--', linewidth=1.5, alpha=0.7)

# Mark some key points
ax.plot(1, stats.norm.cdf(1), 'go', markersize=8, label='P(Z ≤ 1) = 0.8413')
ax.plot(-1, stats.norm.cdf(-1), 'mo', markersize=8, label='P(Z ≤ -1) = 0.1587')

ax.set_title('Cumulative Distribution Function (CDF)', fontsize=12, fontweight='bold')
ax.set_xlabel('Z-Score')
ax.set_ylabel('Cumulative Probability')
ax.legend(fontsize=9, loc='upper left')
ax.grid(True, alpha=0.3)
ax.set_ylim(0, 1.05)

plt.tight_layout()
plt.savefig('normal_distributions.png', dpi=150, bbox_inches='tight')
plt.show()

print("✓ Visualization saved")
```

## Section 6: Z-Scores and Standardization

```python
print("\n" + "="*80)
print("Z-SCORES AND STANDARDIZATION")
print("="*80)

# Example
mu = 100
sigma = 15
values = [70, 85, 100, 115, 130]

print(f"\nExample: IQ Scores")
print(f"Population Mean (μ) = {mu}")
print(f"Population Std Dev (σ) = {sigma}")
print(f"\nConverting individual scores to Z-scores:")
print()
print(f"{'Original Score':<18} {'Calculation':<40} {'Z-Score':<12} {'Interpretation':<30}")
print("-" * 100)

for val in values:
    z = (val - mu) / sigma
    if z == 0:
        interp = "Equals mean"
    elif z > 0:
        interp = f"{z:.2f} std dev above mean"
    else:
        interp = f"{abs(z):.2f} std dev below mean"
    
    calc = f"({val} - {mu}) / {sigma}"
    print(f"{val:<18} {calc:<40} {z:<12.2f} {interp:<30}")

print()
print("="*80)
print("\nKey Insight:")
print("Z-score tells us how many standard deviations a value is from the mean.")
print("\nUsing Z-scores:")
print("• Can compare values from different distributions")
print("• Can identify outliers (typically Z > 3 or Z < -3)")
print("• Can find probabilities using standard normal tables")
```

## Section 7: Working with Probabilities

```python
print("\n" + "="*80)
print("CALCULATING PROBABILITIES WITH NORMAL DISTRIBUTION")
print("="*80)

# Example: Heights of students
mu_height = 170  # cm
sigma_height = 10  # cm

print(f"\nExample: Height Distribution of Students")
print(f"Mean Height (μ) = {mu_height} cm")
print(f"Standard Deviation (σ) = {sigma_height} cm")
print()

# Problem 1: P(X < 180)
print("1. What is the probability a student is shorter than 180 cm?")
x1 = 180
z1 = (x1 - mu_height) / sigma_height
p1 = stats.norm.cdf(z1, 0, 1)  # Using standard normal
print(f"   Z = ({x1} - {mu_height}) / {sigma_height} = {z1:.2f}")
print(f"   P(X < {x1}) = P(Z < {z1:.2f}) = {p1:.4f} = {p1*100:.2f}%")
print()

# Problem 2: P(X > 185)
print("2. What is the probability a student is taller than 185 cm?")
x2 = 185
z2 = (x2 - mu_height) / sigma_height
p2 = 1 - stats.norm.cdf(z2, 0, 1)  # Upper tail
print(f"   Z = ({x2} - {mu_height}) / {sigma_height} = {z2:.2f}")
print(f"   P(X > {x2}) = P(Z > {z2:.2f}) = {p2:.4f} = {p2*100:.2f}%")
print()

# Problem 3: P(160 < X < 180)
print("3. What is the probability a student's height is between 160 and 180 cm?")
x3_low = 160
x3_high = 180
z3_low = (x3_low - mu_height) / sigma_height
z3_high = (x3_high - mu_height) / sigma_height
p3 = stats.norm.cdf(z3_high, 0, 1) - stats.norm.cdf(z3_low, 0, 1)
print(f"   Z_low = ({x3_low} - {mu_height}) / {sigma_height} = {z3_low:.2f}")
print(f"   Z_high = ({x3_high} - {mu_height}) / {sigma_height} = {z3_high:.2f}")
print(f"   P({x3_low} < X < {x3_high}) = P({z3_low:.2f} < Z < {z3_high:.2f}) = {p3:.4f} = {p3*100:.2f}%")
print()

print("="*80)
```

## Section 8: Visualizing Probability Calculations

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

# Plot 1: P(Z < 1.5)
ax = axes[0, 0]
ax.plot(x, y, 'b-', linewidth=2.5)
x_fill = x[x <= 1.5]
y_fill = stats.norm.pdf(x_fill, 0, 1)
ax.fill_between(x_fill, y_fill, alpha=0.3, color='green')
ax.axvline(1.5, color='red', linestyle='--', linewidth=2)
p = stats.norm.cdf(1.5)
ax.set_title(f'P(Z < 1.5) = {p:.4f} ({p*100:.2f}%)', fontsize=12, fontweight='bold')
ax.set_xlabel('Z-Score')
ax.set_ylabel('Probability Density')
ax.grid(True, alpha=0.3)
ax.set_xlim(-4, 4)

# Plot 2: P(Z > 1.0)
ax = axes[0, 1]
ax.plot(x, y, 'b-', linewidth=2.5)
x_fill = x[x >= 1.0]
y_fill = stats.norm.pdf(x_fill, 0, 1)
ax.fill_between(x_fill, y_fill, alpha=0.3, color='red')
ax.axvline(1.0, color='red', linestyle='--', linewidth=2)
p = 1 - stats.norm.cdf(1.0)
ax.set_title(f'P(Z > 1.0) = {p:.4f} ({p*100:.2f}%)', fontsize=12, fontweight='bold')
ax.set_xlabel('Z-Score')
ax.set_ylabel('Probability Density')
ax.grid(True, alpha=0.3)
ax.set_xlim(-4, 4)

# Plot 3: P(-1 < Z < 1)
ax = axes[1, 0]
ax.plot(x, y, 'b-', linewidth=2.5)
x_fill = x[(x >= -1) & (x <= 1)]
y_fill = stats.norm.pdf(x_fill, 0, 1)
ax.fill_between(x_fill, y_fill, alpha=0.3, color='orange')
ax.axvline(-1, color='red', linestyle='--', linewidth=2)
ax.axvline(1, color='red', linestyle='--', linewidth=2)
p = stats.norm.cdf(1) - stats.norm.cdf(-1)
ax.set_title(f'P(-1 < Z < 1) = {p:.4f} ({p*100:.2f}%)', fontsize=12, fontweight='bold')
ax.set_xlabel('Z-Score')
ax.set_ylabel('Probability Density')
ax.grid(True, alpha=0.3)
ax.set_xlim(-4, 4)

# Plot 4: P(|Z| > 2)
ax = axes[1, 1]
ax.plot(x, y, 'b-', linewidth=2.5)
x_fill_left = x[x <= -2]
y_fill_left = stats.norm.pdf(x_fill_left, 0, 1)
ax.fill_between(x_fill_left, y_fill_left, alpha=0.3, color='purple')
x_fill_right = x[x >= 2]
y_fill_right = stats.norm.pdf(x_fill_right, 0, 1)
ax.fill_between(x_fill_right, y_fill_right, alpha=0.3, color='purple')
ax.axvline(-2, color='red', linestyle='--', linewidth=2)
ax.axvline(2, color='red', linestyle='--', linewidth=2)
p = 2 * (1 - stats.norm.cdf(2))
ax.set_title(f'P(|Z| > 2) = P(Z < -2 or Z > 2) = {p:.4f} ({p*100:.2f}%)', fontsize=11, fontweight='bold')
ax.set_xlabel('Z-Score')
ax.set_ylabel('Probability Density')
ax.grid(True, alpha=0.3)
ax.set_xlim(-4, 4)

plt.tight_layout()
plt.savefig('probability_calculations.png', dpi=150, bbox_inches='tight')
plt.show()

print("✓ Visualization saved")
```

## Final Summary

### Key Formulas:

**Normal Distribution PDF:**
$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

**Z-Score (Standardization):**
$$Z = \frac{X - \mu}{\sigma}$$

**Empirical Rule:**
- ±1σ: 68% of data
- ±2σ: 95% of data
- ±3σ: 99.7% of data

### Properties:
1. Symmetric around mean (μ)
2. Bell-shaped curve
3. Mean = Median = Mode
4. Total area under curve = 1 (100%)
5. Asymptotic (never touches x-axis)

### When to Use:
- Modeling continuous data
- Hypothesis testing
- Confidence intervals
- Quality control
- Any data that appears bell-shaped

### Important: Central Limit Theorem
If you take samples from ANY distribution and calculate their means, the distribution of those sample means will be approximately normal, especially as sample size increases. This is why normal distribution is so important!
