# Calculus for Optimization — Chapter 2: Limits


## 2.1 Introduction
Limits describe the behaviour of a function as the input approaches a particular value. They form the foundation for continuity, differentiability, and gradient-based optimization methods used in machine learning.

A limit answers the question:
“What value does the function approach as the input gets arbitrarily close to a specific point?”



## 2.2 Definition of a Limit

$$
\lim_{x \to a} f(x) = L
$$


if the values of \( f(x) \) get arbitrarily close to \( L \) as \( x \) approaches \( a \).

image.png



### Example 1: Polynomial Function
Evaluate:
$$
\lim_{x \to 2} (3x^2 - x + 5).
$$
Since polynomials are continuous:
$$
= 3(2)^2 - 2 + 5 = 15.
$$

## 2.3  

Left-hand limit:
$$
\lim_{x \to a^-} f(x)
$$


Right-hand limit:
$$
\lim_{x \to a^+} f(x)
$$


A limit exists if:
$$
\lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) = L
$$

#### Example 2: Removable Discontinuity
Evaluate:
$$
\lim_{x \to 3} \frac{x^2 - 9}{x - 3}.
$$
Factor:
$$
x^2 - 9 = (x-3)(x+3)
$$
Thus for \(x\neq3\):
$$
\frac{x^2 - 9}{x - 3} = x+3
$$
So:
$$
\lim_{x\to3}(x+3)=6.
$$

#### Example 3: Jump Discontinuity
Given:
$$
f(x)=\begin{cases}
2x-1, & x<1\\
3, & x\ge1
\end{cases}
$$
Left limit:
$$
\lim_{x\to1^-}(2x-1)=1
$$
Right limit:
$$
\lim_{x\to1^+}3=3
$$
They differ, so limit does not exist.

### Example: Infinite Discontinuity at $x = 0$

Consider:

$$
f(x) = \frac{1}{x}
$$


[Image of the graph of f(x) = 1/x, showing the vertical asymptote at x=0]


We examine the behaviour of the function as $x$ approaches $0$.

**Left-hand limit:**

$$
\lim_{x \to 0^-} f(x) = -\infty
$$

**Right-hand limit:**

$$
\lim_{x \to 0^+} f(x) = +\infty
$$

Since the one-sided limits diverge in opposite directions, the limit does not exist in the usual sense.
This type of divergence is called an **infinite discontinuity** (a vertical asymptote).



## 2.5 Algebra of Limits

If
$$
\lim_{x\to a} f(x)=L,\quad \lim_{x\to a} g(x)=M
$$


Then,

Sum:
$$
\lim_{x\to a}(f+g)=L+M
$$


Product:
$$
\lim_{x\to a} f g = LM
$$


Quotient:
$$
\lim_{x\to a} \frac{f}{g} = \frac{L}{M},\ M\neq 0
$$




## 2.6 Standard Limits Useful in Optimization

$$
\lim_{x\to 0} \frac{\sin x}{x} = 1
$$


$$
\lim_{x\to 0} \frac{e^x-1}{x} = 1
$$


$$
\lim_{x\to 0} \frac{\log(1+x)}{x} = 1
$$


These appear in derivatives of exponential, logarithmic, and activation functions.



## 2.7. Why This Matters: The Derivative
The entire concept of "Learning" in AI is based on changing weights to reduce error. We calculate this change using the **Derivative**.

The Derivative is defined **using a limit**:

$$
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
$$

* We calculate the error at $x$.
* We nudge the weight by a tiny amount $h$.
* We check the difference.
* We shrink $h$ toward 0 to find the exact slope.

**Without limits, we cannot define derivatives. Without derivatives, we cannot perform Backpropagation.**





## 2.9 Numerical Demonstration of a Limit

Let's use Python to numerically prove that $\lim_{x \to 0} \frac{\sin x}{x} = 1$.

```python
import numpy as np
import pandas as pd

def f(x):
    return np.sin(x) / x

# Create an array of values getting smaller and smaller (approaching 0)
# We start from 1.0 and go down to 0.00001
values = np.array([1.0, 0.5, 0.1, 0.01, 0.001, 0.0001, 0.00001])

# Calculate f(x) for these values
results = f(values)

# Display in a clean table
df = pd.DataFrame({
    'Input (x)': values,
    'Output (sin x / x)': results
})

print("Approaching 0 from the positive side:")
print(df)

# Note how the Output settles exactly at 1.0 as x gets smaller.
```

## 2.10 Visualization of a Limit

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,400)
y = np.sin(x)/x
y[x==0] = 1

plt.figure(figsize=(7,5))
plt.plot(x,y)
plt.axvline(0, color='black', linestyle='--')
plt.title("Plot of sin(x)/x near x=0")
plt.grid(True)
plt.show()
```

### **Why Continuity Is Required for Differentiability**

To say that a function \( f(x) \) is *differentiable* at a point \( x = a \), we use the definition:

$$
f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}
$$

For this limit to exist, two essential conditions must be satisfied:

---

###### **1. The function value \( f(a) \) must exist**

If the function has a hole, a jump, or goes to infinity at \(x = a\),  
then \( f(a) \) is not defined, and the derivative cannot be computed.

---

##### **2. The function must satisfy**

$$
\lim_{h \to 0} f(a+h) = f(a)
$$

This is **exactly the definition of continuity** at \( x = a \).

Therefore, the function must be continuous at the point to ensure the derivative definition works.

---

##### **Final Rule**

$$
\text{If a function is differentiable at } a,\ \text{then it must be continuous at } a.
$$

But the reverse is not always true:

A function may be continuous but **not differentiable**  
(e.g., ReLU has a sharp corner at \(x = 0\)).

Compute:

\lim_{x\to2}(x^2-5x+6)
$$

### Q2
Does the limit exist?
$$
f(x)=\begin{cases}
x+2,& x<0\\[6pt]
5,& x\ge0
\end{cases}
$$

### Q3
Compute the limit:
$$
\lim_{x\to0}\frac{\sin(3x)}{x}
$$


# Solutions

### A1
Factor:
$$
x^2 - 5x + 6 = (x-2)(x-3)
$$
Substitute \(x = 2\):
$$
(2 - 2)(2 - 3) = 0
$$
**Final Answer:**  
$$
\boxed{0}
$$

---

### A2
Left-hand limit:
$$
\lim_{x\to0^-}(x+2) = 2
$$

Right-hand limit:
$$
\lim_{x\to0^+} 5 = 5
$$

Since:
$$
2 \ne 5
$$

The left and right limits are not equal.

**Final Answer:**  
$$
\boxed{\text{The limit does not exist.}}
$$

---

### A3
Rewrite:
$$
\frac{\sin(3x)}{x}
= 3 \cdot \frac{\sin(3x)}{3x}
$$

Using the standard limit  
$$
\lim_{u\to0}\frac{\sin u}{u} = 1,
$$

we get:
$$
\lim_{x\to0}\frac{\sin(3x)}{x}
= 3 \cdot 1 = 3
$$

**Final Answer:**  
$$
\boxed{3}
$$
