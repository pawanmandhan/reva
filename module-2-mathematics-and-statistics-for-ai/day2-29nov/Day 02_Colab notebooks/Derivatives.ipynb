```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU function
def f(x):
    return np.maximum(0, x)

# Define the derivative of the ReLU function
def df(x):
    # The derivative of ReLU is 1 for x > 0, and 0 for x <= 0
    return np.where(x > 0, 1, 0)

# Generate x values
x_values = np.linspace(-5, 5, 400)

# Calculate y values for the function and its derivative
y_values = f(x_values)
y_prime_values = df(x_values)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='ReLU function: y = max(0, x)', color='blue')
plt.plot(x_values, y_prime_values, label='Derivative of ReLU: dy/dx = 1 (x>0), 0 (x<=0)', color='red', linestyle='--')

# Add labels and title
plt.title('Plot of ReLU Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y / dy/dx')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the Sigmoid function
def dsigmoid(x):
    s = sigmoid(x)
    return s * (1 - s)

# Generate x values
x_values = np.linspace(-5, 5, 400)

# Calculate y values for the function and its derivative
y_sigmoid = sigmoid(x_values)
y_dsigmoid = dsigmoid(x_values)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_sigmoid, label='Sigmoid function: y = 1 / (1 + exp(-x))', color='blue')
plt.plot(x_values, y_dsigmoid, label='Derivative of Sigmoid: dy/dx = sigmoid(x) * (1 - sigmoid(x))', color='red', linestyle='--')

# Add labels and title
plt.title('Plot of Sigmoid Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y / dy/dx')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Tanh function
def tanh(x):
    return np.tanh(x)

# Define the derivative of the Tanh function
def dtanh(x):
    s = tanh(x)
    return 1 - s**2

# Generate x values
x_values = np.linspace(-5, 5, 400)

# Calculate y values for the function and its derivative
y_tanh = tanh(x_values)
y_dtanh = dtanh(x_values)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_tanh, label='Tanh function: y = tanh(x)', color='blue')
plt.plot(x_values, y_dtanh, label='Derivative of Tanh: dy/dx = 1 - tanh(x)^2', color='red', linestyle='--')

# Add labels and title
plt.title('Plot of Tanh Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y / dy/dx')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import sympy as sp

# 1. Define the symbol 'x'
x = sp.Symbol('x')

# 2. Define the function
f = x**2 + 2*x

# 3. Calculate the derivative symbolically
derivative = sp.diff(f, x)

print(f"The formula for the derivative is: {derivative}")

# 4. Calculate the value at a specific point (e.g., x=3)
value_at_3 = derivative.subs(x, 3)
print(f"The value at x=3 is: {value_at_3}")
```
