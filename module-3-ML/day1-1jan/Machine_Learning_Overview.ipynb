## Supervised vs Unsupervised Learning

Supervised and unsupervised learning are two major branches of machine learning.

### Supervised Learning
- **Definition**: Learning with labeled data. The model learns to map **inputs (features)** to **outputs (labels)**.
- **Common Tasks**:
  - üìà **Regression** ‚Äì Predicting continuous values (e.g., house prices)
  - üßÆ **Classification** ‚Äì Predicting categories (e.g., spam or not spam)

  

### üõ†Ô∏è Popular Models:
| Task | Models |
|------|--------|
| Regression | Linear Regression, Decision Tree Regressor, Random Forest Regressor, XGBoost, SVR |
| Classification | Logistic Regression, Decision Tree Classifier, Random Forest Classifier, SVM, KNN, Naive Bayes, Neural Networks |

---

### Unsupervised Learning
- **Definition**: Learning from **unlabeled data**. The model tries to find hidden patterns or structure.
- **Common Tasks**:
  - üìä **Clustering** ‚Äì Grouping similar data points (e.g., customer segments)
  - üîª **Dimensionality Reduction** ‚Äì Reducing the number of variables (e.g., PCA for visualization)

---

### üõ†Ô∏è Popular Models:
| Task | Models |
|------|--------|
| Clustering | K-Means, DBSCAN, Agglomerative Clustering, Gaussian Mixture Models |
| Dimensionality Reduction | PCA, t-SNE, UMAP, Autoencoders |

---

## üîç Other Learning Types (Optional)

| Type | Description | Example |
|------|-------------|---------|
| **Semi-Supervised Learning** | Mix of labeled and unlabeled data | Pseudo-labeling |
| **Reinforcement Learning** | Learn from rewards through exploration | AlphaGo, Q-Learning |
| **Self-Supervised Learning** | Learn representations from raw data | BERT, SimCLR |

---

## Features and Labels

### üîç Features
- Also called: **Input variables**, **Independent variables**, **Predictors**, **X values**
- What the model **learns from**.
- Examples: Age, income, text input, pixels in an image

### üéØ Labels
- Also called: **Target variables**, **Output variables**, **Dependent variables**, **Y values**
- What the model **predicts**.
- Examples: Price of house, disease diagnosis, product category

```python
from graphviz import Digraph

# Create tree structure
dot = Digraph(comment='Types of Learning', format='png')
dot.attr(size='10')

# Root node - Machine Learning
dot.node('L', 'Machine Learning', style='filled', fillcolor='lightblue', shape='box')

# First-level branches
dot.node('SL', 'Supervised Learning', style='filled', fillcolor='lightgreen', shape='box')
dot.node('UL', 'Unsupervised Learning', style='filled', fillcolor='lightyellow', shape='box')
dot.node('SSL', 'Semi-supervised Learning', style='filled', fillcolor='lightsalmon', shape='box')
dot.node('RL', 'Reinforcement Learning', style='filled', fillcolor='lightpink', shape='box')

# Supervised Learning subtypes
dot.node('REG', 'Regression', style='filled', fillcolor='palegreen', shape='ellipse')
dot.node('CLS', 'Classification', style='filled', fillcolor='mediumseagreen', shape='ellipse')

# Unsupervised Learning subtypes
dot.node('CLUST', 'Clustering', style='filled', fillcolor='khaki', shape='ellipse')
dot.node('DR', 'Dimensionality Reduction', style='filled', fillcolor='gold', shape='ellipse')

# Connect nodes
dot.edges([('L', 'SL'), ('L', 'UL'), ('L', 'SSL'), ('L', 'RL')])
dot.edges([('SL', 'REG'), ('SL', 'CLS')])
dot.edges([('UL', 'CLUST'), ('UL', 'DR')])

# Render diagram (for use in Google Colab)
dot.render('/mnt/data/Learning_Tree_Colored', view=False)
dot

```
