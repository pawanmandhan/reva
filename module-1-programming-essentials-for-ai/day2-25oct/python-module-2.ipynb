## Python Variables

```python

# Python Variables

# A variable is created when you assign a value to it.
x = 5
y = "John"

print(x)
print(y)

# You can change the type of the variable after it has been set
x = 4       # x is of type int
x = "Sally" # x is now of type str
print(x)
```

## Variable Names

```python

# Variable Names

# Variable names must start with a letter or the underscore character
myvar = "John"
_my_var = "Jane"
myVar2 = "Alice"

# Variable names are case-sensitive
x = 10
X = 20
print(x)
print(X)

# Invalid variable names (will cause errors if uncommented)
# 2myvar = "John"  
# my-var = "John"
# my var = "John"
```

## Assign Multiple Values

```python

# Assign Multiple Values

# Python allows you to assign values to multiple variables in one line
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)

# One value to multiple variables
x = y = z = "Apple"
print(x)
print(y)
print(z)
```

## Output Variables

```python

# Output Variables

x = "awesome"
print("Python is " + x)

x = "Python is "
y = "awesome"
z = x + y
print(z)

# Using commas adds a space automatically
x = 5
y = 10
print(x, y)
```

## Global Variables

```python

# Global Variables

x = "awesome"

def myfunc():
    print("Python is " + x)

myfunc()

# If you create a variable with the same name inside a function, it will be local
x = "awesome"

def myfunc():
    x = "fantastic"
    print("Python is " + x)

myfunc()

print("Python is " + x)

# Use the global keyword to make a variable global inside a function
def myfunc():
    global x
    x = "super"
myfunc()
print("Python is " + x)
```

## Python Strings

```python

# Python Strings

# Strings in python are surrounded by either single quotation marks, or double quotation marks.
print("Hello")
print('Hello')

# Assign string to a variable
a = "Hello"
print(a)

# Multiline string
b = """This is
a multiline
string."""
print(b)
```

## Slicing Strings

```python

# Slicing Strings

a = "Hello, World!"
print(a[2:5])   # characters from position 2 to 5 (not including 5)
print(a[:5])    # from start to position 5
print(a[2:])    # from position 2 to end
print(a[-5:-2]) # negative indexing
```

## Modify Strings

```python

# Modify Strings

a = " Hello, World! "
print(a.upper())      # Uppercase
print(a.lower())      # Lowercase
print(a.strip())      # Remove whitespace
print(a.replace("H", "J")) # Replace characters
print(a.split(","))   # Split string into a list
```

## Concatenate Strings

## Format Strings

```python

# Format Strings

# We can use f-strings or format() method to combine strings and numbers
age = 25
txt = f"My name is John, and I am {age}"
print(txt)

# Using format() method
quantity = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {:.2f} dollars."
print(myorder.format(quantity, itemno, price))

# Using index numbers
myorder = "I want to pay {2:.2f} dollars for {0} pieces of item {1}."
print(myorder.format(quantity, itemno, price))
```

## Escape Characters

## Python Operators

```python

# Python Operators
# Operators are used to perform operations on variables and values.

# Arithmetic Operators
x = 10
y = 3
print(x + y)  # Addition
print(x - y)  # Subtraction
print(x * y)  # Multiplication
print(x / y)  # Division
print(x % y)  # Modulus
print(x ** y) # Exponentiation
print(x // y) # Floor division

# Assignment Operators
x = 5
x += 3
print(x)

# Comparison Operators
x = 5
y = 10
print(x == y)  # Equal
print(x != y)  # Not equal
print(x > y)   # Greater than
print(x < y)   # Less than

# Logical Operators
a = True
b = False
print(a and b)
print(a or b)
print(not a)

# Identity Operators
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x
print(x is z)       # True
print(x is y)       # False
print(x == y)       # True

# Membership Operators
x = ["apple", "banana"]
print("banana" in x)
print("grape" not in x)

# Bitwise Operators
a = 5  # 0101
b = 3  # 0011
print(a & b)  # AND
print(a | b)  # OR
print(a ^ b)  # XOR
print(~a)     # NOT
print(a << 1) # Left shift
print(a >> 1) # Right shift

# Operator Precedence
x = 5 + 2 * 3
print(x)  # Multiplication has higher precedence
```

## Python Lists

```python

# Python Lists

# A list is a collection which is ordered and changeable.
fruits = ["apple", "banana", "cherry"]
print(fruits)

# Access List Items
print(fruits[1])
print(fruits[-1])
print(fruits[1:3])

# Change List Items
fruits[1] = "blueberry"
print(fruits)

# Add List Items
fruits.append("orange")
print(fruits)

fruits.insert(1, "kiwi")
print(fruits)

# Remove List Items
fruits.remove("apple")
print(fruits)

fruits.pop(1)
print(fruits)

# Loop Lists
for fruit in fruits:
    print(fruit)

# List Comprehension
new_list = [x.upper() for x in fruits]
print(new_list)

# Sort Lists
fruits.sort()
print(fruits)

# Copy Lists
mylist = fruits.copy()
print(mylist)

# Join Lists
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]
list3 = list1 + list2
print(list3)

# List Methods
fruits = ["apple", "banana", "cherry"]
print(len(fruits))
fruits.clear()
print(fruits)

# List Exercises
numbers = [10, 20, 30, 40]
numbers.append(50)
print(numbers)
```

## Python Tuples

```python

# Python Tuples

# A tuple is a collection which is ordered and unchangeable.
thistuple = ("apple", "banana", "cherry")
print(thistuple)

# Access Tuples
print(thistuple[1])
print(thistuple[-1])

# Update Tuples (convert to list, modify, convert back)
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)

# Unpack Tuples
fruits = ("apple", "banana", "cherry")
(green, yellow, red) = fruits
print(green)
print(yellow)
print(red)

# Loop Tuples
for item in fruits:
    print(item)

# Join Tuples
tuple1 = ("a", "b", "c")
tuple2 = (1, 2, 3)
tuple3 = tuple1 + tuple2
print(tuple3)

# Tuple Methods
thistuple = ("apple", "banana", "cherry")
print(len(thistuple))
print(thistuple.count("apple"))
print(thistuple.index("banana"))
```

## Python Sets

```python

# Python Sets

# A set is a collection which is unordered and unindexed.
myset = {"apple", "banana", "cherry"}
print(myset)

# Access Set Items
for x in myset:
    print(x)

# Add Set Items
myset.add("orange")
print(myset)

# Remove Set Items
myset.remove("banana")
print(myset)

# Loop Sets
for x in myset:
    print(x)

# Join Sets
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = set1.union(set2)
print(set3)

# Frozenset (immutable set)
x = frozenset({"apple", "banana", "cherry"})
print(x)

# Set Methods
myset = {"apple", "banana", "cherry"}
print("banana" in myset)
myset.clear()
print(myset)
```

## Python Dictionaries

```python

# Python Dictionaries

# A dictionary is a collection which is unordered, changeable, and indexed.
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)

# Access Items
print(thisdict["model"])
print(thisdict.get("year"))

# Change Items
thisdict["year"] = 2020
print(thisdict)

# Add Items
thisdict["color"] = "red"
print(thisdict)

# Remove Items
thisdict.pop("model")
print(thisdict)

# Loop Dictionaries
for key in thisdict:
    print(key, ":", thisdict[key])

# Copy Dictionaries
mydict = thisdict.copy()
print(mydict)

# Nested Dictionaries
myfamily = {
  "child1" : {"name" : "Emil", "year" : 2004},
  "child2" : {"name" : "Tobias", "year" : 2007}
}
print(myfamily)

# Dictionary Methods
print(thisdict.keys())
print(thisdict.values())
print(thisdict.items())
```
